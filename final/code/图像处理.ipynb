```python
img=cv2.imread("picture/pig.jpg")
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows,cols,_ = img.shape
matrix=cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
img2=cv2.warpAffine(img1,matrix,(cols,rows))
matrix=cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
img3=cv2.warpAffine(img1,matrix,(cols,rows))
matrix=cv2.getRotationMatrix2D((cols/2,rows/2),270,1)
img4=cv2.warpAffine(img1,matrix,(cols,rows))
plt.subplot(221)
plt.imshow(img1)
plt.subplot(222)
plt.imshow(img2)
plt.subplot(223)
plt.imshow(img3)
plt.subplot(224)
plt.imshow(img4)
plt.show()
```

```python
matrix = np.float32([[1,0,25],[0,1,25]])
img2  = cv2.warpAffine(img1,matrix,(cols,rows))
matrix = np.float32([[1,0,50],[0,1,50]])
img3  = cv2.warpAffine(img1,matrix,(cols,rows))
matrix = np.float32([[1,0,75],[0,1,75]])
img4  = cv2.warpAffine(img1,matrix,(cols,rows))
plt.subplot(221)
plt.imshow(img1)
plt.subplot(222)
plt.imshow(img2)
plt.subplot(223)
plt.imshow(img3)
plt.subplot(224)
plt.imshow(img4)
plt.show()
```

```python
SrcPoints = np.float32([[0,0],[0,150],[150,150]])
CanavasPoints = np.float32([[0,0],[0,150],[150,200]])
matrix = cv2.getRotationMatrix2D((0,0),0,0.5)
img2=cv2.warpAffine(img1,matrix,(cols,rows))
matrix = cv2.getAffineTransform(np.array(SrcPoints),np.array(CanavasPoints))
print(matrix)
img3=cv2.warpAffine(img2,matrix,(cols,rows))
plt.subplot(131)
plt.imshow(img1)
plt.subplot(132)
plt.imshow(img2)
plt.subplot(133)
plt.imshow(img3)
plt.show()
```

```python
img = cv2.imread("picture/bigpig.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img1=img.copy()
img2=img.copy()
img1=skimage.util.random_noise(img1,mode="gaussian",seed=None,clip=True,mean=0,var=0.1)
img1=np.uint8(img1*255)
img2=cv2.medianBlur(img1,3)
img1 = np.stack((img1,img1,img1),axis=2)
img2 = np.stack((img2,img2,img2),axis=2)
plt.subplot(121)
plt.imshow(img1)
plt.subplot(122)
plt.imshow(img2)
plt.show()
```

```python
img = cv2.imread("picture/bigpig.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img)
img1=img.copy()
img2=img.copy()
img1=skimage.util.random_noise(img1,mode="gaussian",seed=None,clip=True,mean=0,var=0.1)
img1=np.uint8(img1*255)
img2=cv2.GaussianBlur(img1,(7,7),0)
img3=cv2.GaussianBlur(img1,(19,19),0)
img1 = np.stack((img1,img1,img1),axis=2)
img2 = np.stack((img2,img2,img2),axis=2)
img3 = np.stack((img3,img3,img3),axis=2)
plt.subplot(131)
plt.imshow(img1)
plt.subplot(132)
plt.imshow(img2)
plt.subplot(133)
plt.imshow(img3)
plt.show()
```
